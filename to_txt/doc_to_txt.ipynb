```python
import os
import glob
import textract
```

```python
save_dir = '../../../../data/00_nh_poc/11_legal_advice_hwp_128/text_from_doc/'

for eachfile in glob.glob("../../../../data/00_nh_poc/11_legal_advice_hwp_128/hwp/" + "*.doc"):
    txt = textract.process(eachfile)
                
    file_dir, file_name = os.path.split(eachfile)
    file_name_wo_ext = file_name.split('.doc')[0]
    print(txt)
    
    with open(os.path.join(save_dir, file_name_wo_ext + '.txt'), 'wb') as f:
        f.write(txt)
```
